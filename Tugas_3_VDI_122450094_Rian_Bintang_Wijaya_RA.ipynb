```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/spotify-2023.csv"
data = pd.read_csv(path, encoding='latin-1')
data
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
data.isna().sum()
```

```python
data.info()
```

```python
filtered_song1 = data[data['released_year'] == 2023]
filtered_song1 = filtered_song1[['track_name', 'artist(s)_name', 'released_year', 'streams']]
filtered_song1 = filtered_song1.sort_values(by='streams', ascending=False)
plot_filtered_song1 = filtered_song1.head(10)
filtered_song1
```

```python
plot_filtered_song1['streams'] = pd.to_numeric(plot_filtered_song1['streams'], errors='coerce')

plot_filtered_song1['streams'] = plot_filtered_song1['streams'] / 1000000

plt.figure(figsize=(10, 6))
ax = sns.barplot(y='track_name', x='streams', data=plot_filtered_song1)

plt.title('Top 10 Track Name berdasarkan Streams (in millions)')
plt.ylabel('Track Name')
plt.xlabel('Streams (in millions)')

plt.tight_layout()
plt.show()
```
